## Data wrangling

### Subtask:
Data Wrangling and Feature Engineering


**Reasoning**:
Create new features (profit margin, tax burden, year-over-year growth rates), handle potential errors during calculations, and validate the new features.  Then, aggregate the data to a yearly summary if necessary, and validate the aggregated data.



```python
import pandas as pd
import numpy as np

# Calculate Profit Margin
try:
    df['profit_margin'] = df['laba_bersih_normal'] / df['pendapatan_operasional']
except ZeroDivisionError:
    df['profit_margin'] = np.nan
    print("Warning: ZeroDivisionError encountered while calculating profit margin. Replaced with NaN.")

# Calculate Tax Burden
try:
    df['tax_burden'] = df['pph_badan_normal'] / df['laba_kena_pajak']
except ZeroDivisionError:
    df['tax_burden'] = np.nan
    print("Warning: ZeroDivisionError encountered while calculating tax burden. Replaced with NaN.")

# Calculate Year-over-Year Growth Rates
for col in ['pendapatan_operasional', 'laba_bersih_normal']:
    df[f'{col}_yoy_growth'] = df[col].pct_change() * 100

# Handle potential errors in the new columns (e.g., infinite values)
df = df.replace([np.inf, -np.inf], np.nan)

# Data Validation
print(df.info())
print(df.describe())

#Check for negative profit margin
if (df['profit_margin'] < 0).any():
    print("Warning: Negative profit margin detected.")

# Aggregate data to yearly summary (assuming 'tahun' represents the year, and each row is a month)
df_aggregated = df.groupby('tahun').agg({
    'pendapatan_operasional': 'sum',
    'laba_bersih_normal': 'sum',
    'profit_margin': 'mean',  # Or other appropriate aggregation method
    'tax_burden': 'mean',
    'pendapatan_operasional_yoy_growth': 'mean',
    'laba_bersih_normal_yoy_growth': 'mean'
}).reset_index()

#Validate aggregated data
print(df_aggregated.info())
print(df_aggregated.describe())
display(df_aggregated)
```

## Data visualization

### Subtask:
Visualize key financial trends and metrics from the yearly aggregated data.


**Reasoning**:
Visualize the financial trends and metrics using line and bar charts as requested in the instructions.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set Seaborn style
sns.set_style("whitegrid")

# Convert 'tahun' to string for plotting if it's not already
df_aggregated['tahun_str'] = df_aggregated['tahun'].astype(str)

# Plot 1: Line chart for pendapatan_operasional and laba_bersih_normal over time
try:
    plt.figure(figsize=(12, 6))
    sns.lineplot(data=df_aggregated, x='tahun_str', y='pendapatan_operasional', marker='o', label='Pendapatan Operasional')
    sns.lineplot(data=df_aggregated, x='tahun_str', y='laba_bersih_normal', marker='o', label='Laba Bersih Normal')
    plt.title('Pendapatan Operasional and Laba Bersih Normal Over Time')
    plt.xlabel('Tahun')
    plt.ylabel('Amount')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.savefig('pendapatan_laba_trend.png')
    plt.show()
except Exception as e:
    print(f"Error plotting Pendapatan Operasional and Laba Bersih Normal: {e}")

# Plot 2: Bar chart for average profit_margin and tax_burden across years
try:
    df_melted = df_aggregated.melt(id_vars='tahun_str', value_vars=['profit_margin', 'tax_burden'], var_name='Metric', value_name='Value')
    plt.figure(figsize=(12, 6))
    sns.barplot(data=df_melted, x='tahun_str', y='Value', hue='Metric', palette='viridis')
    plt.title('Average Profit Margin and Tax Burden Across Years')
    plt.xlabel('Tahun')
    plt.ylabel('Average Value')
    plt.legend(title='Metric')
    plt.grid(axis='y')
    plt.tight_layout()
    plt.savefig('profit_tax_comparison.png')
    plt.show()
except Exception as e:
    print(f"Error plotting Profit Margin and Tax Burden: {e}")


# Plot 3: Line chart for year-over-year growth rates
try:
    plt.figure(figsize=(12, 6))
    sns.lineplot(data=df_aggregated, x='tahun_str', y='pendapatan_operasional_yoy_growth', marker='o', label='Pendapatan Operasional YoY Growth')
    sns.lineplot(data=df_aggregated, x='tahun_str', y='laba_bersih_normal_yoy_growth', marker='o', label='Laba Bersih Normal YoY Growth')
    plt.title('Year-over-Year Growth Rates')
    plt.xlabel('Tahun')
    plt.ylabel('Growth Rate (%)')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.savefig('yoy_growth_rates.png')
    plt.show()
except Exception as e:
    print(f"Error plotting Year-over-Year Growth Rates: {e}")
```
